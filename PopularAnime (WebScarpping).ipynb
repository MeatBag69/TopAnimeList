```python
import requests
```

```python
#this will make the requests from the website
url =' https://myanimelist.net/topanime.php?type=bypopularity'
result = requests.get(url)
```

```python
result.text
```

```python
#part of a beautiful soup library
import bs4
```

```python
#this will create soup for the object
soup = bs4.BeautifulSoup(result.text, 'lxml')
```

```python
#this will find all the ranks from the site
rank = soup.find_all('td', class_ = 'rank ac')
```

```python
rank
```

```python
#this will find all names from the site
title = soup.find_all('div', class_ = 'detail')
```

```python
title
```

```python
#this will scrape all the score 
score = soup.find_all('td', class_ = 'score ac fs14')
```

```python
score
```

```python
#python list
data = []
titles = []
scores = []
```

```python
#find the span and get data rank for the list
for i in rank:
    span = i .find('span')
    #span = i .find('h3')
    data.append(span.string)
```

```python
data
```

```python
#find the span and get data from title

for i in title:
    span = i .find('h3')
    titles.append(span.string)
```

```python
titles
```

```python
#find the span and get the score from the list
for i in score:
    span = i .find('span')
    scores.append(span.string)
```

```python
scores
```

```python
resultTable = []

for i in range(0, len(titles)):
    resultTable.append([ data[i], titles[i], scores[i] ])
```

```python
#importing panda library so that data can be exported
import pandas as pd
```

```python
#creats a dataframe for the datsets
# al = pd.DataFrame({'Top Anime by Popularity' : data})
al = pd.DataFrame(resultTable, columns=["Rank", "Title", "Score"])
```
